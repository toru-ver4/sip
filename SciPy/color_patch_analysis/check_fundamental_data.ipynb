# Color Patch を無理やりプロットしてみる

## 概要
color patch の xyY データは [ここ](http://www.poynton.com/notes/color/GretagMacbeth-ColorChecker.html) で公開されているので、以下の手順で sRGB色域の RGB値を求める。

1. xyY to XYZ 変換
2. XYZ to RGB 変換

## xyY to XYZ 変換

xyz の導出式より以下のことが成り立つ。

\begin{align}
y &= \frac{Y}{X + Y + Z} \\
\frac{Y}{y} &= X + Y + Z \\
\end{align}

ここで x について変形して以下を得る。

\begin{align}
x &= \frac{X}{X + Y + Z} \\
X &= x(X + Y + Z) \\
X &= x \cdot \frac{Y}{y}
\end{align}

同様に z についても求めると以下が得られる。Y は xyY の値をそのまま使う。

\begin{align}
X &= x \cdot \frac{Y}{y} \\
Z &= z \cdot \frac{Y}{y} \\
\end{align}
